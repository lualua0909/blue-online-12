```python
n = int(input())
aList = input()
```

```python
aList = aList.split()
```

```python
if n == 1:
  if int(aList[0]) == 0:
    print('NO')
  else:
    print('YES')
else:
  # count all 1
  count = sum(1 for i in aList if int(i) == 1)
  if count == n - 1:
    print('YES')
  else:
    print('NO') 
```
