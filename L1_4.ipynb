```python
output = ''
flag = True

for c in range(len(str1)):
  A = str1[c]
  B = str2[c]

  indexA = abc.index(A)
  if flag == True:
    if indexA == 25:
      output += 'a'  
    else:
      output += abc[indexA + 1]
      flag = False
  else:
    output += A 

if (output == str1) or (output == str2):
  output = "No such string"
else:
  output = output[::-1]
print(output)
```
